# Valid Anagram

Question: Given two strings s and t, return true if t is an anagram of s, and false otherwise.

LeetCode link: https://leetcode.com/problems/valid-anagram/


## Algorithm
Describe the approach, pattern, and steps to solve the problem here.


```python
# Algorithm snippet

# Target is to return true if it is an anagram or return false

# sort -> 's' string , 't' string

# sort(s) == sort(t) = true

# I can also return False is length doesn't match)
```

## Solution
Write the final solution here.


```python
# Solution snippet

# Brute Force

def isAnagram(s: str, t: str) -> bool:
    if len(s) != len(t):
        return False
    return sorted(s) == sorted(t)


s = 'anagram'
t = 'nagaram'
print(isAnagram(s, t))

s1 = 'rat'
t1 = 'car'
print(isAnagram(s1, t1))
```
