# 3Sum

Question: Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] that sum to zero without duplicate triplets.

LeetCode link: https://leetcode.com/problems/3sum/


## Algorithm
Describe the approach, pattern, and steps to solve the problem here.


```python
# Algorithm snippet

# Sort the array nums first

# loop from i = 0 to n-2

# Two pointer approach is correct

# left, right =. left = i+1, right = n-1

# if nums[i] > 0 nums is already sorted smallest should be < 0 if positive solution no possible

# skip duplicate numbers

# if found == 0 append the triplet
```

## Solution
Write the final solution here.


```python
# Solution snippet

def threeSum(nums):
    nums.sort()
    n = len(nums)
    res = []

    for i in range(n-2):
        if nums[i] > 0:
            break

        if i > 0 and nums[i] == nums[i-1]:
            continue

        left = i+1
        right = n-1

        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total == 0:
                res.append([nums[i], nums[left], nums[right]])
                left += 1
                right -= 1

                while left < right and nums[left] == nums[left-1]:
                    left += 1

                while left < right and nums[right] == nums[right+1]:
                    right -= 1

            elif total < 0:
                left += 1

            else:
                right -= 1

    return res

print(threeSum([-4,-1,-1,0,1,2]))
```
