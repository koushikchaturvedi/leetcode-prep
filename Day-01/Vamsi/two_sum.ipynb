# Two Sum

Question: Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

LeetCode link: https://leetcode.com/problems/two-sum/


## Algorithm
Describe the approach, pattern, and steps to solve the problem here.


```python
# Algorithm snippet

# Brute Force O(n^2)

# Check every pair to match to target

# 1. Start with for loop initial element is going to be "i" 
# 2. Then for loop with another variable "j" will help me understand i + j = target. (Note: i != j)
# 3. target is to come up with nums[i] + nums[j] = target then we can return the index i, j
# 4. As it's two for loops its O(n^2)

# Follow-up: Can you come up with an algorithm that is less than O(n2) time complexity?

# Optimized Solution - O(n)

# If we can organize nums in data structure it'll be easy to track the index, we can trade space or time as time is priority

# 1. Create a hash map to store indexes
# 2. will difererence diff = target - nums[i]
# 3. if different is in the map we can return map[diff], i
# 4. for hash map we have diff = index saved
```

## Solution
Write the final solution here.


```python
# Brute Force
def solution(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if (nums[i] + nums[j] == target):
                return i,j

nums = [1,4,5,6,7,8,9]
target = 5
print(solution(nums,target))
print("Above numbers should equal: 0,1")


nums1 = [1,4,5,6,7,8,10]
target1 = 10

print(solution(nums1,target1))
print("Above numbers should equal: 1,3")
```

```python
# Optimized Solution
def solution(nums, target):
    map = {}
    for i in range(len(nums)):
        difference = target - nums[i]
        if difference in map:
            return map[difference], i

        map[nums[i]] = i

nums = [1,4,5,6,7,8,9]
target = 5
print(solution(nums,target))
print("Above numbers should equal: 0,1")


nums1 = [1,4,5,6,7,8,10]
target1 = 10

print(solution(nums1,target1))
print("Above numbers should equal: 1,3")
```
